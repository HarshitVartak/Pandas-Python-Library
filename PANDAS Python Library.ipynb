#  Applyimg Function

```python
# import

import pandas as pd
```

```python
# dataframe

df = pd.DataFrame({'Employee ID':[111, 222, 333, 444],
                   'Employee Name':['Chanel', 'steve', 'Mitch', 'Bird'],
                   'Salary [$/h]':[35, 29, 38, 20],
                   'Years of Experience':[3, 4 ,9, 1]})
```

```python
def salary_raise(y):
    return y+2
```

```python
df["New Salary"]=df['Salary [$/h]'].apply(salary_raise)    # adding column
```

```python
df["Emp len"]=df['Employee Name'].apply(len)    # adding column
```

```python
df
```

```python
df.columns    # getting columns names
```

```python
df = df [['Employee ID', 'Employee Name', 'Salary [$/h]', 'Years of Experience']]    # getting sorted columns data
```

```python
df
```

```python
def check(x):            # define function is else
    if x.islower():
        return "yes"
    else:
        return "No"
```

```python
df["check"] = df['Employee Name'].apply(check)    # checking lowercase using column name
df
```

```python
df.sort_values(by='Years of Experience',ascending=False)    # sorting by ascending order using column name
```

```python
df[df['Salary [$/h]']>20]    # getting result of employees whose salary is more then 20
```
